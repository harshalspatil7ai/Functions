```python
"""function is the block of code which perform specific task .which helps organising resusing code 
and improve reddbility"""

def num_even_odd(num):    #parameter is imp
    if num%2==0:
        print("no. is even")
    else:
        print("no. is odd")

num_even_odd(16)
```

```python
#function with multiple parameter
def add(a,b):
    return a+b,a*b
add(2,4)

```

```python
#Default Parameter    #this is not default parmeter it is general
def greet(name):
    print(f"Welcome {name}")
greet("Harshal") 
```

```python
#this is default parameter
def greet(name="Vaishnavi"):
    print(f"Welcome {name}")
greet() 
```

```python
#positioal argument
def print_number(*Harshal):  #this is because we dont know how many values provide by user
    for number in Harshal:
        print(number)
print_number(1,2,3,4,"vaishnavi")
```

```python
#keyword argument
def key_word(**kwargs): #special for dectionary
    for key,value in kwargs.items():
        print(f"{key}:{value}")
key_word(name="Harshal",surname ="patil",qualification="MSc stat")
```

```python
def print_details(*args,**kwrgs):
    for val in args:
        print(f"positional argument:{val}")
    for key,value in kwrgs.items():
        print(f"{key}:{value}")
print_details(1,2,3,name="harshal",village="Vadgaon")
```
