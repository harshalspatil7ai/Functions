```python
number=[1,2,3,4,5]
list(map(lambda x:x**2,number)) #this is map function
```

```python
numbers1=[1,2,3]
numbers2=[4,5,6]
added_no=list(map(lambda x,y:x+y,numbers1,numbers2))
print(added_no)
```

```python
# map use to convert str into int
list1=["2","4","6","8"]
updated_list=list(map(int,list1))
print(updated_list)
```

```python
list2=['harshal','sunny','chandu']
updated=list(map(str.upper,list2))
print(updated)
```

```python
list2=['HARSHAL', 'SUNNY', 'CHANDU']
updated=list(map(str.lower,list2))
print(updated)
```

```python
def get_name(person):
    return person['name']

people=[
    {'name':'Krish','age':32},
    {'name':'Jack','age':33}
]
list(map(get_name,people))



```
