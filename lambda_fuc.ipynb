```python
#Lambda function are small anonymous function defined using lambda keyword .they have have no. of arguments but only one expression
addition=lambda a,b:a+b
print(type(addition))
print(addition(2,5))
```

```python
multiplication=lambda x,y,z:x*y+z
print(multiplication(2,3,6))
```

```python
even=lambda num:num%2==0
print(even(13))
```

```python
number=[1,2,3,4,5]
list(map(lambda x:x**2,number)) #this is map function
```
